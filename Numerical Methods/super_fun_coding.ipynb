```python
import numpy as np
import matplotlib.pyplot as plt
```

## Basis of attraction

```python
# For (x-4)(x+1)
# Roots = 4, -1

def generate_f1():

  def f1(x):
    return (x-4) * (x+1)
  def df1(x):
    return (2 * x) - 3

  def newton_raphson(x0, max_iter=100):
    for i in range(max_iter):
      x0 = x0 - (f1(x0)/df1(x0))
      if x0 == 4 or x0 == -1:
        return x0
    return x0
  
  def get_color(root):
    if root == 4:
      return 'r'
    else:
      return 'g'
  
  x_range = np.linspace(-2, 5, 100)
  
  plt.plot(x_range, f1(x_range), color='black', linestyle='--')
  
  for x_start in x_range:
    root = newton_raphson(x_start)
    color = get_color(root)
    plt.scatter(x_start, 0, color=color, s=5, marker='.')

  for root in [-1, 4]:
    plt.scatter(root, 0, color=get_color(root), marker='o', edgecolors='black', s=50)

  plt.title("Basins of Attraction for f(x) = (x-4)(x+1)")
  plt.grid()
  plt.show()

generate_f1()
```

```python
# For (x-1)(x+3)
# Roots = 1, -3

def generate_g():

  def g(x):
    return (x-1) * (x+3)
  def dg(x):
    return (2 * x) +2

  def newton_raphson(x0, max_iter=100):
    for i in range(max_iter):
      x0 = x0 - (g(x0)/dg(x0))
      if x0 == 1 or x0 == -3:
        return x0
    return x0
  
  def get_color(root):
    if root == 1:
      return 'r'
    else:
      return 'g'
  
  x_range = np.linspace(-4, 2, 100)

  plt.plot(x_range, g(x_range), color='black', linestyle='--')

  for x_start in x_range:
    root = newton_raphson(x_start)
    color = get_color(root)
    plt.scatter(x_start, 0, color=color, s=5, marker='.')

  for root in [-3, 1]:
    plt.scatter(root, 0, color=get_color(root), marker='o', edgecolors='black', s=50)

  plt.title("Basins of Attraction for f(x) = (x-1)(x+3)")
  plt.grid()
  plt.show()

generate_g()
```

```python
# For (x-4)(x-1)(x+3)
# Roots = 4, 1, -3

def generate_h():

  def h(x):
    return (x-4)*(x-1)*(x+3)
  def dh(x):
    return (3 * (x**2)) - (4 * x) - 11

  def newton_raphson(x0, max_iter=100):
    for i in range(max_iter):
      x0 = x0 - (h(x0)/dh(x0))
      if x0==4 or x0==1 or x0==-3:
        return x0
    return x0
  
  def get_color(root):
    if root == 1:
      return 'r'
    elif root == 4:
      return 'g'
    elif root == -3:
      return 'b'
    else:
      return 'k'

  plt.figure(figsize=(12, 6))
  x_range = np.linspace(-4, 5, 100)

  plt.plot(x_range, h(x_range), color='black', linestyle='--')

  for x_start in x_range:
    root = newton_raphson(x_start)
    color = get_color(root)
    plt.scatter(x_start, 0, color=color, s=19, marker='.')

  for root in [4, 1, -3]:
    plt.scatter(root, 0, color=get_color(root), marker='o', edgecolors='black', s=80)

  plt.title("Basins of Attraction for f(x) = (x-4)(x-1)(x+3)")
  plt.grid()
  plt.xticks(np.arange(-4, 6, 0.5))
  plt.yticks(np.arange(-30, 40, 10))
  plt.show()

generate_h()
```
